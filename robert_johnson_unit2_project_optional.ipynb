### Part 3: Calculate the average price of an order.

Hint: Examine the data to see if the `quantity` column is relevant to this calculation.

Hint: Think carefully about the simplest way to do this!

```python
data.info()
```

```python

data.loc[:,'item_price'] = (data.loc[:,'item_price'].str.replace('$', '')).astype(float)                  # this will drop the $ out of item price which will allow suming

order_sum = data.loc[:, 'item_price'].sum()                                                              #  defines var for the sum of item price note item price accounts for quanity 

order_count = len(data.loc[:, 'order_id'].value_counts())                                                # defines var for number of orders

avg_price = (order_sum / order_count).round(2)                                                           # defines var for average price as a number rounded to two decemial places 

print('$', avg_price)                                                                                   # adds $ to the price
```
